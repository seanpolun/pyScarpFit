```python
import scipy.io as spio
import matplotlib.pyplot as plt
import main
import os
import numpy as np
```

```python
matfile = "C:\\Users\\sgpq58\\Research_Share\\Post Doctoral Research\\scarp_work\\birdseye_creek\\birdseye_scarps.mat"
matdata = spio.loadmat(matfile)
```

```python
q2_data = matdata['q2prof']
x = q2_data[:, 0]
z = q2_data[:, 1]

q2_scarp = main.Scarp(x, z, name="q2_scarp")
q2_scarp.gen_ss()
q2_scarp.gen_1e()
q2_scarp.aspect = 4
q2_scarp.plot_scarp('ss')

```
