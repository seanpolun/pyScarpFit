```python
import scipy.io as spio
import matplotlib.pyplot as plt
import main
import os
import numpy as np
import uncertainties
```

```python
# matfile = "C:\\Users\\sgpq58\\Research_Share\\Post Doctoral Research\\scarp_work\\ferris_mtns\\michael_profiles.mat"
matfile = "C:\\Users\\sgpq58\\Research_Share\\Post Doctoral Research\\scarp_work\\ferris_mtns\\more_pete_profiles.mat"

matdata = spio.loadmat(matfile)
```

```python
out_dir = "C:\\Users\\sgpq58\\Research_Share\\Post Doctoral Research\\scarp_work\\ferris_mtns\\pywork"
keys = ["cherry_q2", "pete_q3p" , "pete_q1p_2"]
names = ["Cherry Creek Q2", "Pete Creek Q3P", "Pete Creek Q1P"]
for i in range(len(keys)):
    key = keys[i]
    table = matdata[key]
    x = table[:, 0]
    z = table[:, 1]
    I = np.argsort(x)
    x = x[I]
    z = z[I]
    z = np.flip(z)
    x = x - x[0]
    scarp = main.Scarp(x, z, b_fit='dsp', name=names[i])
    scarp.gen_ss()
    scarp.gen_1e()
    scarp.aspect = 1

    ss_out = key + "_ss"
    se_out = key + "_se"

    ss_eps = ss_out + ".eps"
    se_eps = se_out + ".eps"

    ss_png = ss_out + ".png"
    se_png = se_out + ".png"

    ss_eps_path = os.path.join(out_dir, ss_eps)
    se_eps_path = os.path.join(out_dir, se_eps)

    ss_png_path = os.path.join(out_dir, ss_png)
    se_png_path = os.path.join(out_dir, se_png)

    scarp.plot_scarp('ss')
    scarp.save_scarp_fig(ss_eps_path)
    scarp.save_scarp_fig(ss_png_path)

    scarp.plot_scarp('se')
    scarp.save_scarp_fig(se_eps_path)
    scarp.save_scarp_fig(se_png_path)
```

```python
cherry_scarp = scarp
cherry_scarp.midx = 9.576
cherry_scarp.midz = 2322.9459
cherry_scarp.b = np.ones_like(cherry_scarp.b) * 0.1621
cherry_scarp.b1 = uncertainties.ufloat(0.1621, 0.001)
cherry_scarp.b2 = uncertainties.ufloat(0.1621, 0.001)
cherry_scarp.H1 = 0.574
cherry_scarp.Hs = 0.665
cherry_scarp.gen_ss()
cherry_scarp.gen_1e()
cherry_scarp.plot_scarp('ss')
```
